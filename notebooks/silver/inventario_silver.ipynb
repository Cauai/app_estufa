```python
# 1) Ler a tabela bronze
query_bronze = "SELECT * FROM bronze.inventario_estufas_bronze"
df_bronze = pd.read_sql(query_bronze, engine)

print("Bronze shape:", df_bronze.shape)
```

```python
# 2) Criar DataFrame da camada silver a partir da bronze
df_silver = df_bronze.copy()
```

```python
# Coluna global das estufas (id da família no Add-on)
df_silver["artigo_id_global"] = "0107"
```

```python
# Coluna por bloco: 010701, 010702, ..., 010725
# Assumindo que 'bloco' é numérico (1,2,3,...)
df_silver["artigo_id_bloco"] = (
    "0107" + df_silver["bloco"].astype(int).astype(str).str.zfill(2)
)
```

```python
# 3) Gravar na camada silver
tabela_silver = "inventario_estufas_silver"

df_silver.to_sql(
    name=tabela_silver,
    con=engine,
    schema="silver",
    if_exists="replace",   # ou "append" na próxima fase
    index=False
)

print(f"Tabela silver.{tabela_silver} carregada com sucesso! ✅")
```
